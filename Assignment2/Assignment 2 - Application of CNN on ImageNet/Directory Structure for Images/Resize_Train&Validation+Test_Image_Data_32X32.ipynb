```python
import pandas as pd 
val_file = pd.read_csv("N:/ImageNet/tiny-imagenet-200/tiny-imagenet-200/val/val_annotations.csv")
val_file
```

```python
#Creating Folder Structure for Validation Images to copy images as per the Class
int_folder = "N:/ImageNet/Image/Val"    
for i in range(0,200):
    val_img = int_folder + "/" + sub_folders[i]
    if not os.path.exists(val_img):
        os.makedirs(val_img)
        print(str(i)+"Folder Created for Class - "  + sub_folders[i])
        
```

```python
#Copying the Images to Class Folder as per the Val_Annotations files
start = time.time()
b = 0 
for j in range(0,len(val_file)):
    b +=1
    source = "N:/ImageNet/tiny-imagenet-200/tiny-imagenet-200/val/images/" + val_file.iloc[j]['Image']
    destination = int_folder + "/" + val_file.iloc[j]['Class']
    shutil.copy(source, destination)
    print("Coying Image : " + str(b))

end = time.time()
print("Time for Execution: " + str(end-start))
```
